```python
import re
from bs4 import BeautifulSoup
from urllib.request import urlopen
linkList = []
url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20190825&page=5"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
```

```python
soup.find_all('div','tit5')
```

```python
for link in soup.find("div", {"id":"content"}).findAll("a", href=re.compile("^(/movie/point)((?!:).)*$")):
    if 'href' in link.attrs:
        #print('https://movie.naver.com' + link.attrs['href'])
        movieLink = 'https://movie.naver.com' + link.attrs['href']
        linkList.append(movieLink)
        print(movieLink)
#         page = urlopen(movieLink)
#         soup = BeautifulSoup(page, "html.parser")
#         print(soup)
```

## 영화 페이지 속에서 데이터 추출

```python
import re
import csv
from bs4 import BeautifulSoup
from urllib.request import urlopen
mvGenre = []
comp = "/movie/sdb/browsing/bmovie.nhn?genre" # 비교용
for i in range(50):
    f = open('output.csv', 'a', encoding='utf-8', newline='')
    url = linkList[i]
    page = urlopen(url)
    soup = BeautifulSoup(page, "html.parser")

    #영화 이름과 장르 추출
    mvTitle = soup.find("div", {"class": "choice_movie_info"}).find("a").text
    
    for atag in soup.find("div", {"class": "choice_movie_info"}).find("td").findAll("a"):
        if comp in str(atag):
            mvGenre.append(atag.text)
        
    print(mvTitle) #제목 크롤링
    print(mvGenre) #장르 크롤링

    #영화 평점 추출
    stPoint = soup.find("div", {"class":"fl point_type_b"}).find("div")
    stPoint = str(stPoint)
    stPoint = stPoint[31:35]
    stPoint = float(stPoint)/10
    stPoint = round(stPoint, 2)
    print(stPoint)

    #영화 투표수 추출
    soup.findAll("span")
    for i in soup.findAll("td"):
        if i.find("em") != None:
            voteCnt = i.find("em").text
            print(voteCnt)

    # 영화 포스터 추출
    poster = soup.find("div", {"class": "fl"}).find("img")['src']
    print(poster)
    print("")

    #csv 파일에 저장
    wr = csv.writer(f)
    wr.writerow([mvTitle, mvGenre, stPoint, voteCnt, poster])
    del mvGenre[:]
    f.close()
print("총 개수", i)
```
