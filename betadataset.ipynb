### 영화 데이터 추출 프로그램

```python
import re
import csv
from bs4 import BeautifulSoup
from urllib.request import urlopen
linkList = []
mvGenre = []
comp = "/movie/sdb/browsing/bmovie.nhn?genre" # 비교용
base_url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20190825&page={}"
Max_PageNum = input("원하는 페이지 수를 선택하시오 : ")
for j in range(int(Max_PageNum)): # 원하는 페이지 까지 루틴
    url = base_url.format(j+1) #랭킹 페이지 값을 변경
    page = urlopen(url)
    soup = BeautifulSoup(page, "html.parser") # beautifulsoup 링크 페이지를 실행
    #print(soup) 
    print("진행상황",j, "/40", url) # 랭킹 페이지가 변경 되는 것을 확인

    #영화 링크를 가져오는 부분, 이부분에서 문제가 있음을 알 수 있음
    linkList = []
    for link in soup.find("div", {"id":"content"}).findAll("a", href=re.compile("^(/movie/point)((?!:).)*$")):
        if 'href' in link.attrs:   
            #print('https://movie.naver.com' + link.attrs['href'])
            movieLink = 'https://movie.naver.com' + link.attrs['href']
            linkList.append(movieLink)
            #print(movieLink)
    #         page = urlopen(movieLink)
    #         soup = BeautifulSoup(page, "html.parser")
    #         print(soup)
    
    for i in range(50): #가져온 랭킹 페이지의 영화 링크대로 하나하나 추출작업
        f = open('output.csv', 'a', encoding='utf-8', newline='')
        url2 = linkList[i]
        page2 = urlopen(url2)
        soup2 = BeautifulSoup(page2, "html.parser")

        #영화 이름과 장르 추출
        mvTitle = soup2.find("div", {"class": "choice_movie_info"}).find("a").text

        for atag in soup2.find("div", {"class": "choice_movie_info"}).find("td").findAll("a"):
            if comp in str(atag):
                mvGenre.append(atag.text)

        print(mvTitle) #제목 크롤링
        print(mvGenre) #장르 크롤링

        #영화 평점 추출
        stPoint = soup2.find("div", {"class":"fl point_type_b"}).find("div")
        stPoint = str(stPoint)
        stPoint = stPoint[31:35]
        stPoint = float(stPoint)/10
        stPoint = round(stPoint, 2)
        print(stPoint)

        #영화 투표수 추출
        soup2.findAll("span")
        for i in soup2.findAll("td"):
            if i.find("em") != None:
                voteCnt = i.find("em").text
                print(voteCnt)

        # 영화 포스터 추출
        poster = soup2.find("div", {"class": "fl"}).find("img")['src']
        print(poster)
        print("")

        #csv 파일에 저장
        wr = csv.writer(f)
        wr.writerow([mvTitle, mvGenre, stPoint, voteCnt, poster])
        del mvGenre[:]
        f.close()
```
